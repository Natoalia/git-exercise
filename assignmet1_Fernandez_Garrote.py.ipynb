```python
"""Write a python script assignment1_your_lastname.py that reads a DNA sequence from the file ACB04367_dna.txt, translates it to an amino acid sequence with single-letter representations and writes the AA-sequence to a file ACB04367_aa.txt.
The inputfile ACB04367_dna.txt contains a string of DNA, represented by single-letter nucleotides ('A','C','G','T').
The standard.trans text file contains standard translation between DNA nucleotides and single- or three-letter representation of amino acids.
The script could contain a function translate(infile,outfile) that reads-in the nucleotide sequence from an arbitrary input file (infile), translates it to an AA sequence (single letter representation) using standard.trans and writes it to an output file ACB04367_aa.txt. If a more 'pythonian' implementation comes to your mind, go for it!
You can assume that the directory of the script already contains the file standard.trans as well as the ACB04367_dna.txt file, so submit the script only."""


#DICTIONARY
#Firstly, create the dictionary, which will be use to translate 
#between DNA codons to aa, provided by the file standard.trans
    
#open the file with the translation code
transfile=open('standard.trans','r') 
#read the file
transstring=transfile.read() 
#length of the file
D=len(transstring) 

#crete the dictionary
translate = {
    }
   
#Each line of the file is read in each iteration.
    #D-1 because there is an extra blanck character at the end of the file
for k in range(0,(D-1),10):#each line contains 9 characters
    #Get the input codon, 3 first characters of each line
        codon=transstring[k:(k+3)]
    #Get the aa letter the codon codes for, 5th character of each line
        letter=transstring[(k+4)]
    #Set the translation into the dictionary
        translate[codon]=letter

#close the transfile as it won't be further used
transfile.close() 


#READ INPUT FILE
#Now, read the input DNA sequence from the file ACB04367_dna.txt
infile = open('ACB04367_dna.txt','r')
instring=infile.read()
N = len(instring)

#EXCEPTION HANDLING
#First of all we need the chain to have n bases, n being divisible by 3, as codons are groups of 3 bases
if N%3==1:
    print("Error: length of the sequence is not divisible by 3")
    exit()
 

#OUTPUT  
#Creating the output file
outfile=open('ACB04367_aa.txt','w')

#Every codon will be read individually
for i in range(0,N,3):
    try:
        #each codon is 3 letters, it is translated with the dictionary just created
        out=translate[instring[i:(i+3)]]
        #rewrite the outfile
        outfile.write(out)
    except: #EXCEPTION HANDLING if the codon does not have a translation
        print("Error: codon does not codify")
        exit()
        #this part is not necessary, as every combination codes for a aa
        #could be useful for other dictionaries

#close the files
infile.close()
outfile.close()

#these lines were used to check the script
#poo=open('ACB04367_aa.txt','r')
#roo=poo.read()
#print(roo)
    
```
